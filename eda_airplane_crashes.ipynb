EXPLORATORY DATA ANALYSIS ON AIRPLANE CRASHES DATASET

The goal of this task is to explore and understand the historical airplane crash dataset through statistical summaries and data visualizations. By analyzing trends, distributions, and relationships between features, we aim to uncover patterns or anomalies that may provide valuable insights into aviation safety and accident history.

STEP 1: IMPORT REQUIRED LIBRARIES


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

%matplotlib inline

sns.set(style='whitegrid')
```

STEP 2: LOAD THE DATSET

```python
# Set the correct file path to load the dataset
file_path = r"C:\Users\uthil\Desktop\INTERNSHIP\TASK 2\airplane_crash dataset.csv"

# Load the dataset
df = pd.read_csv(file_path)
df.head()  
```

STEP 3: BASIC DATA EXPLORATION

Understand Data Structure

```python
df.info()

df.describe()

df.columns

df.describe(include='all')
```

Check for Missing Values

```python
df.isnull().sum()
df.isnull().mean() * 100 
```

 Summary Statistics

```python
df.describe(include='all') 
```

STEP 4: VISUALIZE DISTRIBUTIONS OF NUMERIC FEATURES

Visualize the distributions of some important numeric features (e.g., Fatalities, Aircraft Age).

```python
# Histogram of Fatalities
plt.figure(figsize=(8,4))
sns.histplot(df['Fatalities'].dropna(), bins=30, kde=True)
plt.title('Distribution of Fatalities')

# Boxplot for Fatalities
plt.figure(figsize=(8,4))
sns.boxplot(x=df['Fatalities'])
plt.title('Boxplot of Fatalities')

# Boxplot for Aircraft Age (if present in the dataset)
if 'Aircraft Age' in df.columns:
    plt.figure(figsize=(8,4))
    sns.boxplot(x=df['Aircraft Age'])
    plt.title('Boxplot of Aircraft Age')
```

STEP 5: CORRELATION MATRIX (FOR NUMERIC FEATURES)

Visualize the correlation between numeric features using a heatmap. This will help you identify if any features are strongly correlated

```python
df_numeric = df.select_dtypes(include=[np.number])

plt.figure(figsize=(10,8))
sns.heatmap(df_numeric.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
```

STEP 6: TREND OVER TIME (YEAR-WISE ANALYSIS)

If dataset contains a Date or Year column, you can analyze trends over time, like the number of crashes per year.

```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')  # Use `errors='coerce'` to handle any bad data
df['Year'] = df['Date'].dt.year

crashes_per_year = df['Year'].value_counts().sort_index()

plt.figure(figsize=(12,5))
crashes_per_year.plot()
plt.title('Crashes per Year')
plt.xlabel('Year')
plt.ylabel('Number of Crashes')
```

STEP 7: IDENTIFY OUTLIERS

It is identified using boxplots

```python
# Boxplot for Fatalities
plt.figure(figsize=(8,4))
sns.boxplot(x=df['Fatalities'])
plt.title('Boxplot of Fatalities')

# Boxplot for Aircraft Age (if present)
if 'Aircraft Age' in df.columns:
    plt.figure(figsize=(8,4))
    sns.boxplot(x=df['Aircraft Age'])
    plt.title('Boxplot of Aircraft Age')
```

STEP 8: PAIRPLOT FOR RELATIONSHIS

```python
sns.pairplot(df[['Fatalities', 'Aboard', 'Ground']])
```

STEP 9: INTERACTIVE VISUALIZATIONS

```python
if 'Aircraft Age' in df.columns:
    fig = px.scatter(df, x='Aircraft Age', y='Fatalities', title='Aircraft Age vs Fatalities')
    fig.show()
```
